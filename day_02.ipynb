# 表达式
- 由一个或几个数字或者变量或运算符合成的一行代码
- 通常返回一个结果

## 运算符
- 有一个以上的值经过一系列的运算得到新值的过程叫运算
- 用来操作运算的符号叫做运算符
- 运算符分类：
    - 算术运算符
    - 比较或者关系运算符
    - 赋值运算符
    - 逻辑运算符
    - 位运算符
    - 成员运算符
    - 身份运算符

```python
#表达式 案例
a = 1 + 2

```

## 算术运算符
- 用来计算算术的符号
- 通常用来表示加减乘除
- Python没有自增自减运算符

```python
# 算术运算符案例

# 加减乘除数学意义基本一致
a = 9 - 2
print(a)

b = 9 + 2
print(b)

c = 9 * 2 
print(c)

# python 除法分为通常除法 /; 地板除（取整） // ; 取余 %
d = 9 / 2
print(d)

e = 9 // 2 #取整
print(e)

# 取余公式：q = c*h + r  q 为被除数  c 为商  h:除数 r:余数  0 <= |r| < |h|
f = 9 % 2
print(f)

i = 9 % 5
print(i)

g = -9 % 5
print(g)

h = 9 % -5
print(h)

#指数
m = 7 ** 3
print(m)
```

# 比较运算符
- 对两个内容进行比较的运算符
- 结果一定是布尔值，即True或False


```python
# 等于 == 
a = 3 == 4
print(a)

# 不等于 !=
b = 3 != 4
print(b)

#其他符号 > >= < <=
c = 4 >= 4
print(c)

d = 5 <= 4
print(d)

f = 5.0 == 5.0000000000
print(f)
```

# 赋值运算符
- 把一个值放到变量里面去
    

```python
# 赋值符号 =
a = 9

# 复杂赋值
a = b = 9
a, b = 1, 3
print(a, b)

#赋值的缩写
## 所有的数学运算符都可以进行缩写
# -= += /= //= %= **= *= 都是缩写形式
cc = 100
cc = cc + 3
print(cc)

cc += 3  # xx += b 等同于 xx = xx + b
print(cc)

d = 9 
d **= 2  # 指数运算缩写
print(d)
```

# 逻辑运算符
- 对布尔类型变量或者值进行运算的符号
- and ： 逻辑与
- or： 逻辑或
- not：逻辑非
- Python里面的逻辑运算没有异或


## 运算规则
- and看做乘法   or 看做加法
- True 看做1 False看做0，则逻辑运算就能转换成整数数学运算，最后结果如果是0则为False 否者为True

## 逻辑运算的短路问题
-  已经知道了结果，即不往下继续运算(从左向右)


```python
# 逻辑表达式举例 
a = True
b = False
c = False

aa = a and b
print(aa)

cc = a or c
print(cc)

dd = 100 and c 
print(dd)

ff = not a
print(ff)

if not c:
    print("打印")
```

```python
# 短路问题案例
a = True
b = True
c = False
aa = a or c and (b and c) #已经知道了结果，即不往下继续运算(从左向右)
print(aa)
```

```python
#短路问题案例2
def a():
    print("a")
    return True

def b():
    print("b")
    return False
cc = a() and b()
print(cc)
print("*" * 20)

dd = a() or b()
print(dd)
print("-" * 20)
```

# 成员运算符
 - 用来检查一个值或者变量是否在某个集合里面
 - in：成员运算符
 - not in：不在里面的意思
 

```python
# in 案例
l = [1, 2, 3, 4, 5]
a = 6
aa = a in l
print(aa)

# a 没在成语里面
bb = a not in l
print(bb)
```

# 身份运算符
- 用来确定两个变量是否是同一变量
- is：变量运算符
- is not：不是

```python
#身份运算符定义
a = 1
b = 199934349
aa = a is b
print(aa)

# a,b 仅仅是值一样，并不代表a，b是一个变量
a = 10000000
b = 10000000
bb = a is b
print(bb)

#正确理解下面的案例
# 特别注意 小一点的数字，当值在 [-5,256]之间时,在Python解释器中为同内存区块
a = 256
b = 256
cc = a is b
print(cc)
```
