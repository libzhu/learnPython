# 转义字符
- 用一个特色的方法表示出一系列不方便写出的内容，比如回车键、换行符， 退格键
- 借助反斜杠\字符， 一旦字符串出现反斜杠字符，则反斜杠后面的一个或者几个字符表示已经不是原来的意思了，进行了转义
- 在字符串中，一旦出现反斜杠就要加倍小心，可能由转义字符出现
- 不同系统对换行有不同的表示
    - window：\n
    - Linux: \r\n
    
- 转义字符	

        \(在行尾时)	续行符
        \\	反斜杠符号
        \’	单引号
        \”	双引号
        \a	响铃
        \b	退格(Backspace)
        \e	转义
        \000	空
        \n	换行
        \v	纵向制表符
        \t	横向制表符
        \r	回车
        \f	换页
        \oyy	八进制数yy代表的字符，例如：\o12代表换行
        \xyy	十进制数yy代表的字符，例如：\x0a代表换行
        \other	其它的字符以普通格式输出

```python
# 转义字符案例
# 想表达 Let‘s Go
# 使用转义字符
s = 'Let\'s Go'
print(s)

# 使用单双引号嵌套
s = "Let's Go"
print(s)

# 表示斜杠
# 比如表示 C:\User\Vteam
s = "C:\\User\\Vteam"
print(s)

#  回车换行
# 表达想需要的效果
s = "I\nlove\nBeiJing"
print(s)
```

# 格式化
- 把字符串按照一定的格式打印或填充
    - 格式化的分类：
            换说明符	说明

            %d，%i	转换为带符号的十进制形式的整数
            %o	转换为带符号的八进制形式的整数
            %x，%X	转换为带符号的十六进制形式的整数
            %e	转化为科学计数法表示的浮点数（e 小写）
            %E	转化为科学计数法表示的浮点数（E 大写）
            %f，%F	转化为十进制形式的浮点数
            %g	智能选择使用 %f 或 %e 格式
            %G	智能选择使用 %F 或 %E 格式
            %c	格式化字符及其 ASCII 码
            %r	使用 repr() 将变量或表达式转换为字符串
            %s	使用 str() 将变量或表达式转换为字符串
    


```python
# 填充
s = "I love BeiJing"
print(s)

# %s 表示简单的字符串
# 占位符可以单独使用
s = "I love %s"
print(s)

print(s%"WuHan")
print(s%"ShenZhen")

# 一般占位符只能被同类型的值替换，
print("I Love %s"%"ShangHai")
```

```python
s = "小明，今年%d岁了"
# print(s%"19")  #报错 类型错误
print(s%19)
```

```python
s = "你的体重：%fKG, 身高：%fm"
print(s)
print(s%(66.4, 175.33))


s = "你的体重：%.2fKG, 身高：%.2fm"
print(s)
print(s%(66.4, 175.33))

s = "你的体重：%gKG, 身高：%gm"
print(s)
print(s%(70.3, 170.3))

```

# format 格式化
- 使用函数的格式进行格式化， 代替以前的百分号


```python
# 不指定位置， 按顺序读取

# 方式一
s = "{} {}!"
print(s)
print(s.format("hello", "world"))

# 方式二
s = "{} {}".format("hello", "world!")
print(s)

# 方式三
# 指定位置
s = "{0} {1}".format("hello", "world!")
print(s)

s = "{1} {0}".format("hello", "world!")
print(s)

s = "{1} {1}".format("hello", "world!")
print(s)

s = "I Love {0} and {0} Love Me".format("BeiJing")
print(s)

# 下面案例报错，和上面对比
# s = "I Love {} and {} Love Me".format("BeiJing")
# print(s)

# 使用命名参数
s = "我们是{companyName}公司, 我们的网址是：{web}, {name}最帅"
s = s.format(companyName="天逸", web="www.vteamsystem.com", name="小明")
print(s)
```

```python
# 通过字典设置参数 需要解包
s = "我们是{companyName}公司, 我们的网址是：{web}, {name}最帅"
s_dict = {"companyName" : "天逸",
         "web" : "www.vteamsystem.com",
         "name" : "小朱"}
# ** 是解包操作
s = s.format(**s_dict)
print(s)
```

```python
# 对数字的格式化需要用到
s = "我的身高 {:.2f}m height, {:.2f}kg weight"
print(s.format(1.74, 79))
```

# str 内置函数
- 字符串查找 find, index islower
- find rfind(从右开始查找): 查找字符串中是否包含一个子串 返回要查找的字符串的开始下标位置或没有找到返回-1
- index: 查找字符串的下标，返回要查找的字符串的开始下标位置, 要是没有找到字符串，会系统报错、引发异常

```python
s = "BeiJing tian an men wo ai li"
s1 = "tian"
# 返回第一次发现这个字符串的位置 从字符串零位开始
s.find(s1)

s2 = "wangwang"
# 返回-1表示没有找到
s.find(s2)
```

```python
s  = "我爱北京天安门"
s1 = "天安门"
print(s.find(s1))
print(s.rfind("天"))
s2 = "上海"
print(s.find(s2))
```

```python
s  = "我爱北京天安门"
s1 = "天安门"
print(s.index(s1))

s2 = "上海"
s.index(s2)

```

```python
# 使用的时候可以使用区间
s = "Liu Dan love Wang xiaojing and Zhang xiaojing"
s1 = "xiaojing"
s2 = "hello"
# 从下标20的位置开始找，看能否找到
s.find(s1, 20)
s.find(s2, 20)
```

# 字符串判断类函数
- 此类函数的特点一般是用is开头， 比如 islower
- isalpha:判断是否是字母， 需要注意：
    - 此函数默认的前提是字符串中至少包含一个字符，如果没有同样返回False
    - 汉字被认为是alpha，此函数不能作为区分英语字母和汉字的标识，区分中英文使用Unicode码
- isdigit, isnumeric, isDecimal 三个判断数字的函数
    - 此类函数不建议使用，在后期爬虫中，判断是否是数字建议采用正则表达式
            
            对这三个函数的总结：
                isdigit()：
                True：Unicode数字， byte数字(单字节)，全角数字（双字节），罗马数字
                False：汉字数字
                Error：无

                isdecimal():
                True: Unicode数字， 全角数字（双字节）
                False:罗马数字，汉字数字
                Error：byte数字（单字节）

                isnumberic()
                True：Unicode数字， 全角数字（双字节）罗马数字 汉字数字
                False: 无
                Error：byte数字（单字节）
                

```python
# 以下三个都不是，应为有除了字母以外的空格
s1 = "我们对着灯发誓，牛大拿同学是爱着王晓静的"
s2 = "bengberba is friend of baberbeng"
s3 = "TuLingXueYuan.com is the 1.st"

print(s1.isalpha())
print(s2.isalpha())
print(s3.isalpha())
```

```python
# 需要注意：因为输入法的问题，输入罗马数字可能得不到我们想要的结果
chin_num = "一二三四"
print(chin_num.isdigit())
print(chin_num.isnumeric())
print(chin_num.isdecimal())
```

# 内容判断
- startswith/endswith：是否以xxx开头或结尾
    - 检查字符串是否已某个字符串开头
    - suffix：被检查的字符串
    - start：检查范围的开始
    - end: 检查范围的结尾
- islower/isupper: 判断字符串是否是大写或者小写

```python
# startswith/endswith 案例
dana = "Liu Dana"
xiaojing = "Xiao jing"
s  = "Liu Dana really love Wang Xiao jing"
print(s.startswith(dana))
print(s.endswith(xiaojing))
```

```python
# islower/isupper 案例
s1 = "Dana love wang xiao jing"
s2 = "Danalovewangxiaojing"
s3= "danalovewangxiajing"
# s4 含空格，但空格不影响结果 忽略
s4 = "dana love wangxiaojing"
s5 = "刘大拿同学爱过王晓静" 
s6 = "YYFFNJFFF"

print(s1.islower())
print(s2.islower())
print(s3.islower())
print(s4.islower())
print("*" * 20)
# 汉字字符串无大小写概念 
print(s5.islower()) #False
print(s5.isupper())

print("*" * 20)
# isupper
print(s1.isupper())
print(s6.isupper())
```

# 字符串操作类函数
- format：格式化用的
- strip：主要作用是删除字符串两边的空格，同样还有rstrip和lstrip, 此处l和r分别代表左边和右边，，即删除字符串左边或右边定制的字符串，默认为空格。需要注意的是此处的删除不是删除一个，即从开头开始符合条件的连续字符串
- join: 一个字符串加入到另一个字符串中去

```python
s = " I Love beijing tian an men "
print(s.strip(), end="----")  #删除 左右的空格字符串

print()

print(s.strip(" I "))

print("*" * 20)
```

```python
# join的例子 -- 字符串列表按指定的字符分割成新字符串
s1 = "$"
s2 = "-"
s3 = " "
s4 = ","
ss = ["LiuDana", "love", "Wang", "xiaojing"]
print(s1.join(ss))
print(s2.join(ss))
print(s3.join(ss))
print(s4.join(ss)) 
```
